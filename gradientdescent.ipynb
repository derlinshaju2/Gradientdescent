```python
import streamlit as st
import numpy as np
import pandas as pd
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
```

```python
m=0
b=0
lr=0.01
for i in range(10):
    yp=m*x+b
    dm=(-2/len(x))*sum(x*(y-yp))
    db=(-2/len(x))*sum((y-yp))
    m=m-lr*dm #partial derivative
    b=b-lr*db
    mse=(1/len(x))*sum((y-yp)**2)
    print('m',m,'b',b,'mse',mse)
```
